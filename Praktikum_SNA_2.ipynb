#CENTRALITY MEASURE

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
```

#Degree Centrality

```python
from networkx.algorithms.bipartite.centrality import degree_centrality
G = nx.Graph()
G.add_edges_from([(1,2), (2,3), (2,4), (3,4), (4,5), (5,6)])

degree_centrality = nx.degree_centrality(G)

print(degree_centrality)
```

```python
df = pd.DataFrame({'from': [1,2,2,3,4,4,5],
                    'to': [2,3,4,4,5,6,6]})

G = nx.from_pandas_edgelist(df, source='from', target='to')

degree_centrality = nx.degree_centrality(G)

print(degree_centrality)
```

#Betweeness

```python
df = pd.DataFrame({'from':[1,1,1,2,2,3,3,4,4,5],
                   'to':[2,3,4,3,4,4,5,5,6,6]})

G = nx.from_pandas_edgelist(df, 'from', 'to')

betweenness_centrality = nx.betweenness_centrality(G)

print(betweenness_centrality)
```

```python
df = pd.DataFrame({'from':[1,1,1,2,2,3,3,4,4,5],
                   'to':[2,3,4,3,4,4,5,5,6,6]})

G = nx.from_pandas_edgelist(df, 'from', 'to')

betweenness_centrality = nx.betweenness_centrality(G)

pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos, node_size = 100, node_color =list(betweenness_centrality.values()), cmap=plt.cm.Reds)
nx.draw_networkx_edges(G,pos)
nx.draw_networkx_labels(G,pos)

max_bc = max(betweenness_centrality.values())
high_bc_nodes =[k for k, v in betweenness_centrality.items() if v == max_bc]
nx.draw_networkx_nodes(G,pos, nodelist=high_bc_nodes, node_size=200, node_color ='r')
```

#Closeness Centrality

```python
df = pd.DataFrame({'from': [1, 1, 1, 2, 2, 3, 3, 4, 4, 5], 'to': [2, 3, 4, 3, 4, 4, 5, 5, 6, 6]})

G = nx.from_pandas_edgelist(df, 'from', 'to')

betweenness_centrality = nx.closeness_centrality(G)

print(betweenness_centrality)
```

```python
df = pd.DataFrame({'from': [1, 1, 1, 2, 2, 3, 3, 4, 4, 5], 'to': [2, 3, 4, 3, 4, 4, 5, 5, 6, 6]})

G = nx.from_pandas_edgelist(df, 'from', 'to')

closeness_centrality = nx.closeness_centrality(G)

pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos, node_size=150, node_color=list(closeness_centrality.values()), cmap=plt.cm.Blues)
nx.draw_networkx_edges(G, pos)
nx.draw_networkx_labels(G, pos)
plt.show()
```
