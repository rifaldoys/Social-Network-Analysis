# **Mencari Semua Graph**

```python
#Contoh 5:
#Jalur/ Busur
data_jalur = {'A': ['B', 'C'],'B': ['C', 'D'],'C': ['D'],'D': ['C'],'E': ['F'],'F': ['C']}
#Keterangan: {'A': ['B', 'C'] artinya titik A terhubung dng titik B dan C
```

```python
#Membuat Fungsi untuk menemukan sebuah jalur (path) dari titik awal hingga titik akhir atau tujuan.
def cari_jalur(data_jalur, awal, akhir, jalur=[]):
  jalur = jalur + [awal]
  if awal == akhir:
    return [jalur]
  if awal not in data_jalur:
    return []
  semua_jalur=[]
  for titik in data_jalur[awal]:
    if titik not in jalur:
      jalur2 = cari_jalur(data_jalur, titik, akhir, jalur)
      for jalur_baru in jalur2:
        semua_jalur.append(jalur_baru)
  return semua_jalur
```

```python
cari_jalur(data_jalur, 'A', 'D')
```
