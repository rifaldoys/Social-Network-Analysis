```python
import networkx as nx
import matplotlib.pyplot as plt
# Membuat jaringan dengan beberapa jenis entitas atau modality
G = nx.Graph()
# Menambahkan entitas ke jaringan
G.add_node('user1', modality='user')
G.add_node('user2', modality='user')
G.add_node('group1', modality='group')
G.add_node('group2', modality='group')
# Menambahkan koneksi antara entitas
G.add_edge('user1', 'user2', relation='friend')
G.add_edge('user1', 'group1', relation='member')
G.add_edge('user2', 'group2', relation='member')
# Menentukan posisi entitas pada grafik
pos = nx.spring_layout(G)
# Menentukan warna untuk setiap modality
node_colors = {'user': 'blue', 'group': 'green'}
# Menggambar entitas pada grafik
for node in G.nodes():
 modality = G.nodes[node]['modality']
 nx.draw_networkx_nodes(G, pos, nodelist=[node], node_color=node_colors[modality])
# Menggambar hubungan antara entitas pada grafik
nx.draw_networkx_edges(G, pos)
# Menampilkan label pada entitas
labels = {node: node for node in G.nodes()}
nx.draw_networkx_labels(G, pos, labels)
# Menampilkan grafik
plt.axis('off')
plt.show()
```

```python
import networkx as nx

#buat jaringan dengan beberapa jenis entitas atau modality
G = nx.Graph()

#tammbahkan entitas ke jaringan
G.add_node('Alice', modality='owner')
G.add_node('Bob', modality='owner')
G.add_node('Charlie', modality='owner')
G.add_node('Pizza Hut', modality='business')
G.add_node('Starbucks', modality='business')
G.add_node('McDonalds', modality='business')

#tambahkan koneksi antara entitas
G.add_edge('Alice', 'Pizza Hut', relation='owns')
G.add_edge('Bob', 'Pizza Hut', relation='owns')
G.add_edge('Charlie', 'Starbucks', relation='owns')
G.add_edge('Charlie', 'McDonalds', relation='owns')

#menghitung degree centrality berdasarkan modality
owner_centrality = nx.degree_centrality(G.subgraph((n for n, attr in G.nodes(data=True) if attr['modality'] == 'owner')))
business_centrality = nx.degree_centrality(G.subgraph((n for n, attr in G.nodes(data=True) if attr['modality'] == 'business')))

#menampilkan hasil
print("Degree Centrality for Owners:")
for node, centrality in owner_centrality.items():
    print(f"{node}: {centrality}")

print("\nDegree Centrality for Businesses:")
for node, centrality in business_centrality.items():
    print(f"{node}: {centrality}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
```

```python
# Membaca data
data = pd.read_excel('datarelasi.xlsx')
```

```python
# Membuat graph menggunakan NetworkX
G = nx.Graph()

for index, row in data.iterrows():
    entitas1 = row['Entitas1']
    entitas2 = row['Entitas2']
    relasi = row['Relasi']

    if entitas1 in ['BOSA', 'INFINTY', 'CYBER']:
        G.add_node(entitas1, modality='Organisasi')
    else:
        G.add_node(entitas1, modality='Orang')

    if entitas2 in ['BOSA', 'INFINTY', 'CYBER']:
        G.add_node(entitas2, modality='Organisasi')
    else:
        G.add_node(entitas2, modality='Orang')

    G.add_edge(entitas1, entitas2, relation=relasi)
```

```python
# Menghitung derajat sentralitas (degree centrality)
degree_centrality = nx.degree_centrality(G)

# Menghitung antara sentralitas (betweenness centrality)
betweenness_centrality = nx.betweenness_centrality(G)

# Menghitung jarak sentralitas (closeness centrality)
closeness_centrality = nx.closeness_centrality(G)

# Menghitung derajat kepentingan sentralitas (eigenvector centrality)
eigenvector_centrality = nx.eigenvector_centrality(G, )

# Menampilkan entitas dengan degree centrality tertinggi
highest_degree = max(degree_centrality, key= degree_centrality.get)
print("Entitas dengan degree centrality tertinggi:", highest_degree)

# Menampilkan entitas dengan betweenness centrality tertinggi
highest_betweenness = max(betweenness_centrality, key=betweenness_centrality.get)
print("Entitas dengan betweenness centrality tertinggi:", highest_betweenness)

# Menampilkan entitas dengan closeness centrality tertinggi
highest_closeness = max(closeness_centrality, key=closeness_centrality.get)
print("Entitas dengan closeness centrality tertinggi:", highest_closeness)

# Menampilkan entitas dengan eigenvecotr centrality tertinggi
highest_eigenvector = max(eigenvector_centrality, key=eigenvector_centrality.get)
print("Entitas dengan eigenvector centrality tertinggi:", highest_eigenvector)

```

Interpretasi Hasil:

Entitas dengan degree centrality, betweenness centrality, closeness centrality, dan eigenvector centrality tertinggi itu sama yaitu John. Hal tersebut menandakan bahwa John memiliki peran yang sangat penting dalam jaringan yang terbentuk.

Baik dalam jumlah hubungan yang dimiliki (degree centrality). Baik sebagai penghubung antara entitas yang satu dengan yang lain (Betweenness Centrality), Baik sebagai entitas dengan jarak terdekat ke entitas lain (Closeneess Centrality). Maupun berdasarkan hubungan dengan node-node penting dalam jaringan (Eigen vecotr Centrality)

```python
# Menampilkan grafik jaringan biasa
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G)
nx.draw_networkx(G, pos, with_labels=True, node_color='lightblue', node_size=1000, font_size=10, font_weight='bold')
plt.title("Grafik Jaringan")
plt.show()
```

```python
#Visualisasi Berdasarkan modality dan relasi
plt.figure(figsize=(10, 8))

# Menentukan posisi entitas pada grafik
pos = nx.spring_layout(G, k=0.7)

# Menentukan warna untuk setiap modality
node_colors = {'Orang': 'blue', 'Organisasi': 'green'}

# Menggambar entitas pada grafik
for node in G.nodes():
    modality = G.nodes[node]['modality']
    nx.draw_networkx_nodes(G, pos, nodelist=[node], node_color=node_colors[modality], node_size=1000)

# Menggambar hubungan antara entitas pada grafik
nx.draw_networkx_edges(G, pos)

# Menampilkan label pada entitas
labels = {node: node for node in G.nodes()}
nx.draw_networkx_labels(G, pos, labels, font_size=10, font_color='red')

# Menampilkan label pada hubungan (relasi)
relation_labels = nx.get_edge_attributes(G, 'relation')
nx.draw_networkx_edge_labels(G, pos, edge_labels=relation_labels, font_size=6)

# Menampilkan grafik
plt.title("Grafik Jaringan")
plt.show()
```

#Degree Centrality

```python
# Menghitung degree centrality berdasarkan modality
centrality = nx.degree_centrality(G)

orang_centrality = {node: centrality for node, centrality in centrality.items() if G.nodes[node]['modality'] == 'Orang'}
organisasi_centrality = {node: centrality for node, centrality in centrality.items() if G.nodes[node]['modality'] == 'Organisasi'}

# Mengurutkan hasil degree centrality berdasarkan nilai centrality tertinggi
sorted_orang_centrality = dict(sorted(orang_centrality.items(), key=lambda x: x[1], reverse=True))
sorted_organisasi_centrality = dict(sorted(organisasi_centrality.items(), key=lambda x: x[1], reverse=True))

# Menampilkan hasil degree centrality yang diurutkan
print("Degree Centrality (Orang)")
for node, centrality in sorted_orang_centrality.items():
    print(f"{node}: {centrality}")

print("\nDegree Centrality (Organisasi)")
for node, centrality in sorted_organisasi_centrality.items():
    print(f"{node}: {centrality}")
```

**Interpretasi Hasil:**

Degree centrality adalah analisis jaringan yang mengukur seberapa banyak node (entitas) terhubung dengan node lainnya dalam sebuah jaringan. Sentralitas ini menggambarkan seberapa penting atau berpengaruh sebuah node dalam jaringan berdasarkan jumlah koneksi yang dimiliki.

Dari perhitungan Degree centrality yang dilakukan diperoleh node yang memiliki degree centrality tertinggi adalah **"John"** dengan skor 0.3076 untuk node dengan modality "**Orang**". Sedangkan untuk modality "**Organisasi**" adalah **"BOSA"** dengan skor degree centrality sebesar 0.1538

Sentralitas ini menggambarkan sejauh mana sebuah node berada di antara komunikasi atau perpindahan informasi antara node lain dalam jaringan.

Dari perhitungan Betweenness centrality yang dilakukan diperoleh node yang memiliki betweenness centrality tertinggi adalah **"John"** dengan skor 0.2948 untuk node dengan modality "**Orang**". #Eigenvector Centrality

```python
# Menghitung eigenvector centrality berdasarkan modality
centrality = nx.eigenvector_centrality(G)

orang_centrality = {node: centrality for node, centrality in centrality.items() if G.nodes[node]['modality'] == 'Orang'}
organisasi_centrality = {node: centrality for node, centrality in centrality.items() if G.nodes[node]['modality'] == 'Organisasi'}

# Mengurutkan hasil eigenvector centrality berdasarkan nilai centrality tertinggi
sorted_orang_centrality = dict(sorted(orang_centrality.items(), key=lambda x: x[1], reverse=True))
sorted_organisasi_centrality = dict(sorted(organisasi_centrality.items(), key=lambda x: x[1], reverse=True))

# Menampilkan hasil eigenvector centrality yang diurutkan
print("Eigenvector Centrality (Orang)")
for node, centrality in sorted_orang_centrality.items():
    print(f"{node}: {centrality}")

print("\nEigenvector Centrality (Organisasi)")
for node, centrality in sorted_organisasi_centrality.items():
    print(f"{node}: {centrality}")
```

**Interpretasi Hasil:**

Eigenvector centrality adalah analisis jaringan yang mengukur seberapa berpengaruh sebuah node dalam jaringan berdasarkan seberapa banyak node-node terhubung dengan node tersebut. Sentralitas ini menggambarkan seberapa "penting" sebuah node dalam jaringan dengan memperhitungkan kualitas hubungannya dengan node-node lain.

Dari perhitungan Eigenvector centrality yang dilakukan diperoleh node yang memiliki eigenvector centrality tertinggi adalah **"John"** dengan skor 0.6310 untuk node dengan modality "**Orang**".
